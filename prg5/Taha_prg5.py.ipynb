```python
import csv,random,math
import statistics as st

def loadcsv(filename):
  lines=csv.reader(open(filename,"r"));
  dataset=list(lines)
  for i in range(1,len(dataset)):
    dataset[i]=[float(x) for x in dataset[i]]
  return dataset  


dataset=loadcsv("p5.csv");
print("pima indian diabetes loaded")
print("total instances available:",len(dataset))
print("total attributes prsent:",len(dataset[1])-1)


def splitDataset(dataset,splitRatio):
  testSize=int(len(dataset)*splitRatio);
  trainSet = list(dataset);
  testSet=[]
  while len(testSet) < testSize:
    index=random.randrange(len(trainSet));
    testSet.append(trainSet.pop(index))
  return [trainSet,testSet]  


def separatedByClass(dataset):
  separated ={}
  for i in range(1,len(dataset)):
    x=dataset[i]
    for i in range(1,len(dataset)):
      x = dataset[i]
      if (x[-1] not in separated):
        separated[x[-1]]=[]
      separated[x[-1]].append(x)  
    return separated  


def compute_mean_std(dataset):
  mean_std=[(st.mean(attribute),st.stdev(attribute))
    for attribute in zip(*dataset)];
  del mean_std[-1]
  return mean_std


def summarizedByClass(dataset):
  separated=separatedByClass(dataset);
  summary={  }
  for classValue,instances in separated.items():
    summary[classValue]=compute_mean_std(instances)
  return summary 

def estimateProbability(x,mean,stdev):
  exponent=math.exp(-(math.pow(x-mean,2)/(2*math.pow(stdev,2))))
  return (1/(math.sqrt(2*math.pi)*stdev))*exponent
 

def calculateClassProbabilities(summaries,testVector):
  p={}
  for classValue,classSummaries in summaries.items():
    p[classValue]=1
    for i in range(1,len(classSummaries)):
      mean,stdev=classSummaries[i]
      x=testVector[i]
      p[classValue]*=estimateProbability(x,mean,stdev);
  return p  

def predict(summaries,testVector):
  all_p=calculateClassProbabilities(summaries,testVector)
  print("all_p")
  print(all_p)
  bestLabel,bestProb=None,-1
  for lbl,p in all_p.items():
    if bestLabel is None or p>bestProb:
      bestProb=p

      bestLabel=lbl
  return bestLabel

def perform_classification(summaries,testSet):
  predictions=[]
  for i in range(len(testSet)):
    result=predict(summaries,testSet[i])
    predictions.append(result)
  return predictions  

def getAccuracy(testSet,predicions):
  correct=0
  for i in range (1,len(testSet)):
    if testSet[i][-1]==predictions[i]:
      correct+=1
  return(correct/float(len(testSet)))*100.0
splitRatio=0.2
trainingSet,testSet= splitDataset(dataset,splitRatio) 
print("training datatset is spilt into training and testing dataset") 
print("training example =(0)\n testing examle =(1)".format(len(trainingSet),len(testSet))) 

summaries= summarizedByClass(trainingSet);
print("summaries")
print(summaries)

predictions=perform_classification(summaries,testSet)
print("predictions")
print(predictions)
accuracy=getAccuracy(testSet,predictions)
print("accuracy of the naive bayes classfier is",accuracy)
```
